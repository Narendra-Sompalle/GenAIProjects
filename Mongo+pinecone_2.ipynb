```python
from sentence_transformers import SentenceTransformer, util
```

```python
embedding_model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
def get_relavent_docs(text):
  embedds=embedding_model.encode(text)
  embedds=embedds.tolist()
  return index.query(vector=embedds,top_k=2)
```

```python
user_query="give best horror movie?"
```

```python
relevent_docs=get_relavent_docs(user_query)
```
